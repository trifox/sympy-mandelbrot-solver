```
# imports
from  sympy import *;
from sympy.utilities.lambdify import lambdify
from IPython.display import display
#from sympy.printing.theanocode import theano_function

# define symbol c to solve for 
c=symbols("c")
```

```

# the mandelbrot equation as sympy notation
def mandelbrot(xx):
    return xx**2+c
```

```

# start with z=0 and iterate the equation up to wanted depth 
mandel=mandelbrot(0)
for i in range(4):
    mandel=mandelbrot(mandel)
# asign the full mandel equation to a new variable
eq=Eq(0,mandel)
```

```
# output
print("The Mandelbrot equation to solve would be")
display(eq)
```

```
# solving may easily take ages
solution=solve(eq,c)
# print info in jupyter that step is done
print("solved")
```
